```python

missing_values = ["n/a", "na", "--"]
df = pd.read_csv("G:\project\PRoject final yr\data.csv", na_values = missing_values)
```

```python
cnt=0
for row in df:
    try:
        int(row)
        df.loc[cnt]=np.nan
    except ValueError:
        pass
    cnt+=1
```
